```python
import pandas as pd
import pandas as pd
import scipy.stats
import seaborn as sn
import statsmodels.formula.api as sm
```

```python
VideoGameData = pd.read_excel('video_games.xlsx')
```

```python
VideoGameData.head()
```

```python
model_reviewscore = sm.ols('Sales ~ Review_Score',data=VideoGameData).fit()
print(model_reviewscore.summary())
```

```python
model_reviewscore = sm.ols('Sales ~ Used_Price',data=VideoGameData).fit()
print(model_reviewscore.summary())
```

```python
model_reviewscore = sm.ols('Sales ~ Release_Year',data=VideoGameData).fit()
print(model_reviewscore.summary())
```

```python
model_reviewscore = sm.ols('Sales ~ Length_All_PlayStyles_Polled',data=VideoGameData).fit()
print(model_reviewscore.summary())
```

```python
model_reviewscore = sm.ols('Sales ~ Review_Score + Used_Price + Release_Year + Length_All_PlayStyles_Polled',data=VideoGameData).fit()
print(model_reviewscore.summary())
```

```python
# Simple Review = 24.6 Multiple = 17
# Simple Used_Price = 55.8 Multiple = 46.5
# Simple Release_Year = unsignificant Multiple = Significant and -58.5
# Simple Length = 1.7 Multiple = 1.2
```
