```python
import pandas as pd
import scipy.stats
import seaborn as sn
import statsmodels.formula.api as sm
StudentsPerformance = pd.read_csv('StudentsPerformance.csv')
```

```python
import matplotlib.pyplot as plt
reading_score = StudentsPerformance['reading_score']
math_score = StudentsPerformance['math_score']
writing_score = StudentsPerformance['writing_score']
```

```python
plt.scatter(reading_score,math_score,c=StudentsPerformance['gender'].map({'female':'red','male':'blue'}))
```

```python
plt.scatter(writing_score,reading_score,c='yellow')
plt.xlabel('Writing Score')
plt.ylabel('Reading Score')
```
